```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from datetime import datetime


%matplotlib inline
```

```python
df = pd.read_csv('df_after.csv')
```

## this show that the number of males is greater than the number of females and others

```python
sb.countplot(data=df,x='member_gender');
plt.title('Comparison between genders');
```

## this show that the number of users whose age from 30 to 40 is bigger than any other age stage

```python
#Histogram (age_user)
bins = np.arange(0,df['age_user'].max()+5,5)
plt.hist(data=df,x='age_user',bins=bins);
plt.xlim((20,80));
plt.title('frequency of any age between 20 to 80');
plt.xlabel('Age');
plt.ylabel('count');
```

## this show that Thursday is the busiest day of the week in terms of users, and Saturday and Sunday are the least crowded days of the week

```python
# Bar chart (day) 
sb.countplot(data=df,x='day',color = sb.color_palette()[0],order = (['Mon','Tue','Wed','Thu','Fri','Sat','Sun'])); 
plt.title('frequency of any day a week');
```

## this show that The time from eight to ten and from seventeen to twenty are the two times in which there are most users

```python
#Histogram (hour)
bins = np.arange(0,df['hour'].max()+1,1)
plt.hist(data=df,x='hour',bins=bins);
plt.xlim((0,24));
plt.title('frequency of any hour a day')
plt.xlabel('hour');
plt.ylabel('Count');
```

## this show that The probability of the presence of subscribers is more than the probability of the presence of customers in the age group between twenty and eighty, regardless of the duration sec

```python
# Scatter (Age),(duration_sec),(user_type)
g = sb.FacetGrid(data=df,hue='user_type',hue_order=['Customer','Subscriber'],height=4,aspect=1.5);
g = g.map(sb.regplot,'age_user','duration_sec',x_jitter=.04,fit_reg = False);
g.add_legend();
plt.xlabel('Age');
plt.ylabel('duration_sec');
plt.title('Relation between Age & duration sec and user type');
```
